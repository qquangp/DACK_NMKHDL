```python
import requests
import json
import pandas as pd
```

```python
API_key = '31c4a0771a0556ec808b4c09c73e965e'
link = 'http://api.openweathermap.org/data/2.5/air_pollution/history?'
urls = [f'{link}lat=17.1540&lon=106.6602&start=1622505600&end=1624579200&appid={API_key}',
        
            f'{link}lat=37.6162&lon=55.7540&start=1622505600&end=1630454400&appid={API_key}',
        
            f'{link}lat=30.5922&lon=114.3059&start=1606780800&end=1608422400&appid={API_key}'
      ]
```

```python
aqi_provinces = []
for url in urls:
    json_str = requests.get(url).text
    json_pydata = json.loads(json_str)
    aqi_provinces.append(json_pydata)
```

```python
def split(list):
    new = {}
    new['co']=list['components']['co']
    new['no']=list['components']['no']
    new['no2']=list['components']['no2']
    new['o3']=list['components']['o3']
    new['so2']=list['components']['so2']
    new['pm2_5']=list['components']['pm2_5']
    new['pm10']=list['components']['pm10']
    new['nh3']=list['components']['nh3']
    new['aqi']=list['main']['aqi']
    return new
    
    
```

```python
def remake(data):
    new=[]
    for i in data['list']:
        new.append(split(i))
    return new
    
```

```python
data = []
for prov in aqi_provinces:
    data += remake(prov)
len(data)
```
